```python
import sys; sys.path.append('..')
import config, pandas as pd
from src import clustering
df = pd.read_csv(config.DENGUE_SEMANAL)
df.shape
```

## Estadísticas descriptivas

```python
df['casos'].describe()
```

## Outliers (regla 1.5 × IQR) sobre semanas con casos

```python
pos = df.loc[df['casos']>0,'casos']
q1,q3 = pos.quantile(.25), pos.quantile(.75); iqr=q3-q1
lim = q3 + 1.5*iqr
print('limite superior:', lim)
print('semanas outlier:', int((df['casos']>lim).sum()))
```

## Perfil por distrito y selección de k

```python
perfil = clustering.district_profile(df)
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(perfil[clustering.PERFIL_COLS])
clustering.evaluar_k(X)
```

## K-means final e interpretación de clusters

```python
res = clustering.run_kmeans(perfil)
print('k=', res['k'], '| silueta=', round(res['silueta'],3))
clustering.resumen_clusters(res['perfil'])
```
